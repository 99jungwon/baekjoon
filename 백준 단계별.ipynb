# 2884번

```python
hour, minute = map(int, input().split()) 

if minute < 45: ## 분이 45분보다 작을 때
    if hour == 0:
        hour = 23
        minute += 60
    else:
        hour -= 1
        minute += 60
        
print(hour, minute - 45)
## 분이 45분 보다 크면 그냥 시간은 그대로에 분민 -45를 하면 된다
```

# 2480번

```python
a, b, c = map(int, input().split())
if a == b and b == c:
    print(10000 + (a * 1000))
elif (a == b or b == c):
    print(1000 + (b*100))
elif (a == c or b == c):
    print(1000 + (c*100))
else:
    print(100 * max(a,b,c))
```
